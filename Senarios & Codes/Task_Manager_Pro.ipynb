Scenario:

You're a busy professional trying to stay organized amidst your hectic schedule. To keep track of your tasks efficiently, you've developed a simple task management tool in Python. With this tool, you can add, view, mark as completed, and delete tasks with ease.

As you start your day, you open up your Python program. It presents you with a menu of options:

Add Task: You can input a new task to be added to your task list.
View Task: You can view all the tasks currently on your list.
Mark Task: You can mark a task as completed once you've finished it.
Delete Task: You can remove a task from your list if it's no longer relevant.
Exit: If you're done managing tasks, you can choose to exit the program.
You decide to start by adding your tasks for the day. One by one, you input each task into the program. As you progress through your day, you periodically check your task list to see what needs to be done next. When you complete a task, you mark it as done to keep your list up to date. If you no longer need to do a task, you simply delete it from the list.

This tool helps you stay organized and focused, ensuring that you tackle your tasks efficiently and never miss a beat in your busy schedule.

```python
user_task = []

def menu():
    print("1. 'add task' enter 1: ")
    print("2. 'view task' enter 2: ")
    print("3. 'mark task' enter 3: ")
    print("4. 'delete task' enter 4: ")
    print("5. 'Nothing' enter 5:")

def add_task():
    task1 = input("Enter your task:")
    user_task.append(task1)
    print(f"YOUR TASK SUCCESSFULLY ADD {user_task}")

def view_task():
    print(f"{user_task}")

def mark_task():
    task3 = input("enter task your want to mark:")
    if task3 in user_task:
     index = user_task.index(task3)
     completed_task = user_task.pop(index)
     print(f"YOUR TASK HAS BEEN MARK {completed_task}:")
    else:
        print(f"WE DONT HAVE THIS TASK {task3}")

def delete_task():
    task4 = input("Enter delete task:")
    if task4 in user_task:
     user_task.remove(task4)
     print(f"YOUR TASK HAS BEEN DELETED {user_task}")
    else:
        print(f"WE DONT HAVE THIS TASK {task4}")

def main():
   while True:
        menu()
        choice = input("Enter menu number: ")

        if choice == "1":
            add_task()
        elif choice == "2":
            view_task()
        elif choice == "3":
            mark_task()
        elif choice == "4":
            delete_task()
        elif choice == "5":
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please enter a number from 1 to 5.")

if __name__ == "__main__":
    main()

      
# In this Scenario i make some TO-DO list where user add task, delete task, mark task and view there task.
# first i make empty lish where task add.Then i make user define function were task menu print.
# Then i make another function where i define view task.Then another functon where user mark task.
# Then i define another funcion where user delete task.
# After this i make main def function where i use while loop which return values till user not enter 5.
# Then i ask user to enter their choice from 1 - 5.
# After that i use conditional statement that run choice of user.
# And the end of this programm if __name__ == "__main__": is a conditional statement that executes the code block 
# below it only if the script is being run directly. This is commonly used to define a set of instructions or 
# functions that should only run when the script is executed as the main program, and not when it is 
# imported as a module into another script.
```
